```python
import pandas as pd
```

```python
df = pd.read_csv('aggregateNetwork_1840_1845.txt')#, header=None, names=['source', 'target', 'weight'])
```

```python
nodes = list(set(df['source']) & set(df['target']))
```

```python
nodes = sorted(nodes)
vertex = [{'id' : 'n' + n} for n in nodes]
```

```python
df = df[~df['target'].isin(targets)]
df = df[~df['source'].isin(sources)]
edge= df.to_dict('records')
for i, e in enumerate(edge) :
    e['id'] = 'e'+str(i)
    e['source'] = 'n' + e['source']
    e['target'] = 'n' + e['target']
```

```python
import json
json.dump({'nodes' : vertex, 'edges' : edge}, open('1840-45.json', 'w'))
```

```python
targets = set(df['target']) - set(df['source'])
```
