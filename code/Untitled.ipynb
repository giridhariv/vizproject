```python
import json
```

```python
df = json.load(open('aggregateNetwork_1840_1880_betweenness.json'))
node_dict = {}
```

```python
for a in df['nodes'] :
    a['id'] = 'n' + a['id']
    node_dict[a['numid']] = a['id']
```

```python
for a in df['edges'] :
    a['source'] = node_dict[a['source']]
    a['target'] = node_dict[a['target']]
```

```python
json.dump(df, open('1840-1880_betweenness.json', 'wb'))
```
