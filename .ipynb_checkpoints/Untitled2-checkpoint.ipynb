```python
import pandas as pd
```

```python
df = pd.read_csv('data/joins/1899.csv', names=['A', 'B', 'W'])
```

```python
df2 = df.pivot_table(index='A', columns='B', values='W', fill_value=0)
```

```python
import numpy as np
adj = np.array(df2)


import matplotlib.pyplot as plt
plt.imshow(adj)
plt.show()
```

```python
hist = np.hist(adj.flatten())
```
